# Step 4 Handle Missing Values

### **Note on pandas Warning: SettingWithCopyWarning**

While filling missing values, we may see this warning:

```

SettingWithCopyWarning: A value is trying to be set on a copy of a slice from a DataFrame.

````

- **What it means:** `df_small` was created as a subset of `df`:
  ```python
  df_small = df[key_columns]
````

So it’s a **view**, not a fully independent copy.

* **Why pandas warns:** When we do something like:

  ```python
  df_small['che_pc_usd'] = df_small['che_pc_usd'].fillna(df_small['che_pc_usd'].mean())
  ```

  pandas warns that modifying a view can sometimes affect the original DataFrame in unexpected ways.

* **Solution:** Create a **copy** of the subset before modifying:


`````python
df_small = df[key_columns].copy()
`````

### **Select Key Columns and Create a Copy**

We select the most relevant columns for our analysis:

- `location` → Country name  
- `year` → Year of data  
- `region` → Geographical region  
- `income` → World Bank income group  
- `che_gdp` → Health spending as % of GDP  
- `che_pc_usd` → Health spending per capita in USD  
- `gdp_usd2022_pc` → GDP per capita (2022 USD)

We create a **copy** of the selected columns to safely modify them without affecting the original dataset:

> Using `.copy()` avoids the `SettingWithCopyWarning` when filling missing values or making other changes.

`````python
df_small['che_pc_usd'] = df_small['che_pc_usd'].fillna(df_small['che_pc_usd'].mean())
df_small['che_gdp'] = df_small['che_gdp'].fillna(df_small['che_gdp'].mean())
df_small['gdp_usd2022_pc'] = df_small['gdp_usd2022_pc'].fillna(df_small['gdp_usd2022_pc'].mean())
`````

### **Fill Missing Values**

We handle missing values in the key numeric columns by filling them with the **mean** of each column:

- `che_pc_usd` → Health spending per capita  
- `che_gdp` → Health spending as % of GDP  
- `gdp_usd2022_pc` → GDP per capita (2022 USD)

> This ensures there are **no missing values** in the dataset, making it ready for analysis.


`````python
# Verify no missing values remain
print(df_small.isnull().sum())
`````

# Step 6: Further Analysis and Insights

### Step 6a: Trends Over Time by Region

`````python
# Group data by year and region, then calculate average per capita health spending
region_trends = df_small.groupby(['year', 'region'])['che_pc_usd'].mean().unstack()

# Plot trends
plt.figure(figsize=(12,6))
for region in region_trends.columns:
    plt.plot(region_trends.index, region_trends[region], marker='o', label=region)

plt.title("Healthcare Spending per Capita Over Time by Region")
plt.xlabel("Year")
plt.ylabel("Health Spending per Capita (USD)")
plt.legend()
plt.grid(True)
plt.show()
`````

**Purpose:**  
Visualize how healthcare spending per capita changes over time for each region.

**Explanation:**
- `groupby(['year', 'region'])['che_pc_usd'].mean().unstack()` – Groups the data by year and region, calculates average spending, and reshapes it for plotting.
- Loop through each region to plot its trend over the years.
- `marker='o'` adds points to the lines for clarity.
- `plt.legend()` differentiates the regions.
- The plot shows which regions are increasing healthcare spending faster.

### Step 6b: Correlation Between GDP per Capita and Health Spending


`````python
# Scatter plot of GDP per capita vs Health Spending per Capita
plt.figure(figsize=(10,6))
plt.scatter(df_small['gdp_usd2022_pc'], df_small['che_pc_usd'], alpha=0.6, color='teal')
plt.title("GDP per Capita vs Health Spending per Capita")
plt.xlabel("GDP per Capita (USD, 2022)")
plt.ylabel("Health Spending per Capita (USD)")
plt.grid(True)
plt.show()

# Calculate correlation coefficient
correlation = df_small['gdp_usd2022_pc'].corr(df_small['che_pc_usd'])
print("Correlation between GDP per capita and Health Spending per Capita:", round(correlation, 2))
`````

**Purpose:**  
Check whether richer countries spend more on healthcare per person.

**Explanation:**
- Scatter plot shows the relationship between GDP per capita and health spending per capita.
- `alpha=0.6` makes points semi-transparent to see overlapping points.
- Correlation coefficient (between -1 and 1) quantifies the strength of the relationship.

`````python
def add_life_expectancy_real(df_in, life_csv_path):
    if os.path.exists(life_csv_path):
        life_df = pd.read_csv(life_csv_path)
        # Basic sanity checks / renaming examples (uncomment/adjust if needed):
        # life_df.rename(columns={'Life expectancy': 'life_expectancy',
        #                         'Country Name':'location',
        #                         'Year':'year'}, inplace=True)
        # Keep only needed cols:
        needed = [c for c in ['location','year','life_expectancy'] if c in life_df.columns]
        if set(['location','year','life_expectancy']).issubset(needed):
            merged = df_in.merge(life_df[['location','year','life_expectancy']],
                                 on=['location','year'], how='left')
            return merged, True, "Merged real life expectancy from CSV."
        else:
            return df_in.copy(), False, "CSV found but required columns missing."
    else:
        return df_in.copy(), False, "CSV not found; falling back to synthetic."
`````

`````python
# ---- OPTION B: Create DEMO-ONLY synthetic life expectancy ----
def add_life_expectancy_synthetic(df_in):
    out = df_in.copy()
    # Use log GDP per capita + a small effect of spending to simulate life expectancy
    # This is ONLY for demo purposes and not valid for conclusions.
    gdp = out['gdp_usd2022_pc'].replace(0, np.nan)
    log_gdp = np.log(gdp).fillna(np.log(gdp.mean()))
    # Scale roughly to 45–85 yrs
    synth = 45 + 6*(log_gdp - log_gdp.min())/(log_gdp.max() - log_gdp.min())*5 + \
            0.02*(out['che_pc_usd'] / (out['che_pc_usd'].quantile(0.95)))
    out['life_expectancy'] = synth.clip(45, 85)
    return out
`````

# Step 5: Simple regression to quantify the effect of health spending on life expectancy.

Step 5A: Import libraries and prepare data

`````python
from sklearn.linear_model import LinearRegression
import numpy as np

# X = spending per capita (reshaped as 2D array), Y = life expectancy
X = merged_df[["spending_per_capita"]].values
Y = merged_df["life_expectancy"].values
`````

Step 5B: Fit linear regression

`````python
# Create model
model = LinearRegression()

# Fit model
model.fit(X, Y)

# Coefficients
print(f"Intercept: {model.intercept_}")
print(f"Slope (coefficient): {model.coef_[0]}")
`````

Step 5C: Make predictions and plot regression line

`````python
# Predictions
Y_pred = model.predict(X)

# Scatter + regression line
plt.figure(figsize=(8,6))
plt.scatter(X, Y, alpha=0.5, color="blue")
plt.plot(X, Y_pred, color="red", linewidth=2)
plt.xscale("log")  # optional: log scale for spending
plt.xlabel("Spending per Capita (USD)")
plt.ylabel("Life Expectancy (years)")
plt.title("Linear Regression: Life Expectancy vs Health Spending")
plt.show()
`````

Step 5D: R² score (how well spending explains life expectancy)

`````python
r2 = model.score(X, Y)
print(f"R² score: {r2:.2f}")
`````

An R² score of 0.30 means that about 30% of the variation in life expectancy is explained by health spending per capita alone.


Step 5E: Log-transformed regression

`````python
# Log-transform spending per capita
X_log = np.log1p(merged_df[["spending_per_capita"]].values)  # log(1 + x) to handle zeroes
Y = merged_df["life_expectancy"].values

# Fit linear regression
model_log = LinearRegression()
model_log.fit(X_log, Y)

# R² score
r2_log = model_log.score(X_log, Y)
print(f"R² score (log-transformed): {r2_log:.2f}")

# Predictions
Y_pred_log = model_log.predict(X_log)

# Plot
plt.figure(figsize=(8,6))
plt.scatter(merged_df["spending_per_capita"], Y, alpha=0.5, color="blue")
plt.plot(merged_df["spending_per_capita"], Y_pred_log, color="red", linewidth=2)
plt.xscale("log")
plt.xlabel("Spending per Capita (USD)")
plt.ylabel("Life Expectancy (years)")
plt.title("Log-Transformed Regression: Life Expectancy vs Health Spending")
plt.show()
`````
